# Probability and Information Theory

### Probability vs Information Theory

<p>Probability allows us to make statements that have elements of uncertainity. Information theory, on the other hand, helps us quantify the amount of uncertaininity within such statement. In this case, you can think of the word statement as a distribution. Under normal circumstances, a programmer can safely assume that a CPU will flawlessly execute instructions without any issue regarding determinism. In machine learning, however, the programmer has to deal with many elements of uncertain quantities and sometimes nondeterministic (stochastic) quantities. <p>
    
> Nearly all activities require some ability to reason in the presence of uncertainty. In fact, beyond mathematical statements that are true by definition, it is difficult to think of any proposition that is absolutely true or any event that is absolutely guaranteed to occur.
>
> [Goodfellow-et-al-2016, Deep Learning](https://www.deeplearningbook.org/)
    
Here are three things that could raise uncertainity:
    1. Inherent stochasticity (eg. quantum mechanics)
    2. Incomplete observability (eg. Monty Hall problem, different view points)
    3. Incomplete Modelling (eg. discretization)

<p>It is worth pointing out that there needs to be a balance between having a model that approximately represents reality and having a model that is exact. This balance must take computational efficiency and memory requirements into account. <p>
    
<p>On the one hand you can view probability as something that will help you communicate information regarding the frequency at which things tend to happen. For example, you can conduct an experiment of flipping coins a given number of times and from that you can compute a probability of obtaining heads. On the other hand, probability is used to express degree of belief like in the case when the doctor tells you have a 30% chance of catching flu if you come across your sick friend. These two types of probability are somewhat different. The latter has a qualitative nature while the former has an experimental nature. <p>

<br>
    
The two forms of probability are:
    - Frequentist probability (flipping coins N number of times)
    - Bayesian probability (qualitative levels of uncertainity)


> Bayesian probability is an interpretation of the concept of probability, in which, instead of frequency or propensity of some phenomenon, probability is interpreted as reasonable expectation representing a state of knowledge or as quantification of a personal belief..
>
> [Wikipedia](https://en.wikipedia.org/wiki/Bayesian_probability)

    
Frequentist probability in code:

```python
import matplotlib.pyplot as plt
import numpy as np

def coinFlip(size):
    flips = np.random.randint(0, 2, size=size)
    return flips.mean()
coinFlip = np.frompyfunc(coinFlip, 1, 1)

x = np.arange(1, 500, 1)
y = coinFlip(x)
plt.figure(figsize=(10,4))
plt.xlabel('# of Draws', fontsize=18)
plt.ylabel('Freq H vs T', fontsize=16)
plt.ylim([0,1])
plt.plot(x, y)

plt.show()
```

### Random variables

<p>If you have a variable $x$ that can take a value out of a set of numbers, or rather, if you have an event $e$ that can take one of the possible events within a mini universe, you would call it a random variable. Random variables describes are simply descriptions of states that are possible and are usually coupled with a probability distribution that specifies how likely each of these states are.<p>
   
<br>
    
### Probability distributions
<p>Given a random variable and the possible events this random variable can assume, the mapping that takes the tuple (variable, event) as an input and  outputs the probability of the tuple occuring is called a probability distribution. Since distributions come in two types, probability distributions can also be divided into two.<p>

Types of probability distributions
    - Discrete variables: probability mass function (PMF)
    - Continious variables: probability density function (PDF)
    
For these two types of distributions, a function must meet a certain criteria in order to be cosidered as a PMF or PDF:
    - All possible states of x must be convered in the domain of the function
    - The probability for all x must at least be zero
    - If you sum probabilities of all possibilities, the function must return 1 (normalized)
-> In essence, what this is basically saying is that the universe of all possible events is bounded and fully contained
   
<br>
    
### Joint Probability
    
<p>It is also possible to have probability distributions describing multiple variables, such distributions are called joint probability distributions: <p>

$$P(x=x,y=y)$$
    

    
<br>

### Marginal Probability
<p>The proability distribution describing a subset of the bigger universe is called the marginal probability distribution:<p>
    
$$\forall x \in x, P(x = x) = \sum_y P(x=x, y=y)$$
    
<br>
    
### Conditional probability
    
<p>Suppose you have two related events, if you know one event has happened for sure you can describe the probability of the other event happening with respect your knowledge. This is called conditional probability:<p>

$$P(y=y|x=x) = \frac{P(y=y,x=x)}{P(x=x)}$$
<br>
$$P(A|B) = \frac{P(AnB)}{P(B)}$$
    
> It is important not to confuse conditional probability with computing what would happen if some action were undertaken. The conditional probability that a person from Germany given that they speak German is quite high, but if a randomly selected person is taught to speak German, their country of origin does not change.
>
> [Goodfellow-et-al-2016, Deep Learning](https://www.deeplearningbook.org/)
 
<br>
    

### The Chain Rule of Conditional Probabilities

<p>Just like in the chain rule of calculus, a join probabiliyt over many random variables may be decomposed into conditional distributions over only one variable. This is called the chain rule or the product rule of probability:<p>
    
$$P(x^{(1)},...,x^{(n)}) = P(x^{(1)}) \prod_{i=2}^{n}P(x^{(i)}| x^{(1),...,x^{(i-1)}})  $$
$$P(a,b,c) = P(a|b,c)P(b,c)$$
$$P(b,c) = P(b|c)P(c)$$
$$P(a,b,c) = P(a|b,c)P(b|c)P(c)$$
<br>

### Independence and Conditional Independence

<p>Similar to the noition of linear independence in linear algebra, we can say two random variables x and y are independent if their probability distribution can be expressed as a product of two factors:<p>
    
$$\forall x \in x, y \in y$$
$$p(x=x,y=y) = p(x=x)p(y=y)$$
    
<p>The two variables can also be conditionally independent given another random variable z:<p>
    
$$\forall x \in x, y \in y, z \in z$$
$$p(x=x,y=y|z=z) = p(x=x|z=z)p(y=y|z=z)$$
    
<br>

### Expectation, Variance and Covariance
    
<p>Suppose you have a bunch of events (codified with a function f that turns events into numbers eg. H:= 1, T:=0) paired with probabilities, then the average of that the codifier when x is drawn from P is called the expectation. If you roll a die, with probability $1/6$ for each outcome, and codify the events as follows $\{1,2,3,4,5,6\}$ for the respective outcomes, then the expectation would be:  (1+2+3+4+5+6)/6 = 7/2. Which says that you can $expect$ the given number if you had rolled the die many many times. The expectation of tossing a fair coin would be 0.5 etc. <p>
    
$$\mathbb{E}[f(x)] = \int p(x)f(x) dx$$
    
<br>

<p>In order to give a measure of how much the values of a function of a rondom variable vary as we sample different values from x from its probability distribution, we use a property called $variance$. The variance tells how much the random variable $x$ $varies$ from the expected value. When the variance is low, then the values of $f(x)$ cluster around the expected value. The sqrt of the variance is known as the standard deviation.<p>
    
$$\text{Var}(f(x)) = \mathbb{E}\big[f(x) - \mathbb{E}[f(x)]^2\big]$$
    
<br>

<p>The sense of how much two values are linearly related to each other, as well as the scale of these variables is given by the covariance:<p>
    
$$Cov(f(x), g(x)) = \mathbb{E}\big[ (f(x) - \mathbb{E}[f(x)]) (g(y) - \mathbb{E}[g(y)])\big]$$

For a variable:
$$C = \frac{1}{n-1}\sum_{i=1}^n(X_i - \bar{X})(X_i - \bar{X})^T$$
    
<p>Since the covariance describes the way in which variables are realted, we can say that if the covariance of two variables in non-zero, then they must be dependent on eachother in a linear way. High absolute values for covariance means that the two variables are far from their respective expected values (means) and also that they change a lot; if positive it means both variables tend to take on relatively high values simultaneously; if negative then one tends to take on high values while the other takes on low values. If the covariance between two variables is zero, it does not necessarily mean that they are independent, only that they are $linearly$ independent. Independence is a stronger requirement than zero covariance because independence excludes non-linear relationships. <p>
    
The covariance matrix of a random vector $x \in \mathbb{R}^n$ is an $ n x n$ matrix such that:
$$\text{Cov}(x)_{i,j} = \text{Cov}(x_i, x_j)$$
Where the diagonal of such matrix gives the variance.
$$C = \begin{bmatrix}
\sigma(x,y) & \sigma(x,y)  \\
\sigma(y,x) & \sigma(y,y) 
\end{bmatrix}$$
    
In code:

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (12, 8)

# Normal distributed x and y vector with mean 0 and standard deviation 1
x = np.random.normal(0, 1, 500)
y = np.random.normal(0, 1, 500)
X = np.vstack((x, y)).T

plt.scatter(X[:, 0], X[:, 1])
plt.title('Generated Data')
plt.axis('equal')
```

```python
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
# Covariance
def cov(x, y):
    xbar, ybar = x.mean(), y.mean()
    return np.sum((x - xbar)*(y - ybar))/(len(x) - 1)

# Covariance matrix
def cov_mat(X):
    return np.array([[cov(X[0], X[0]), cov(X[0], X[1])], \
                     [cov(X[1], X[0]), cov(X[1], X[1])]])

# Calculate covariance matrix 
data = cov_mat(X.T) # (or with np.cov(X.T))
data = {'A': data[0],
        'B': data[1]
        }

df = pd.DataFrame(data,columns=['A','B'])

covMatrix = pd.DataFrame.cov(df)
sn.heatmap(covMatrix, annot=True, fmt='g')
plt.show()
```

```python
# Center the matrix at the origin
X = X - np.mean(X, 0)

# Scaling matrix
sx, sy = 0.7, 3.4
Scale = np.array([[sx, 0], [0, sy]])

# Apply scaling matrix to X
Y = X.dot(Scale)

plt.scatter(Y[:, 0], Y[:, 1])
plt.title('Transformed Data')
plt.axis('equal')

# Calculate covariance matrix
cov_mat(Y.T)
```
